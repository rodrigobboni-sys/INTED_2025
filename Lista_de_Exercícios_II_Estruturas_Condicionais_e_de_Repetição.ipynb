```python
#1 – Em um país fictício, o direito ao voto começa aos 16 anos. Crie um programa que receba a idade de uma pessoa e determine, usando apenas uma condição simples if, se ela já pode votar.
idade = int(input("Digite sua Idade: "))
if idade >= 16 and idade < 130:
  print("Você já possui o direito de votar")
else:
  print("Você ainda não possui idade para votar")
```

```python
#2 – Uma escola deseja automatizar a verificação do desempenho dos alunos. Leia duas notas, calcule a média e utilize uma estrutura if…else para informar se o aluno está aprovado (média ≥ 7) ou reprovado.
nota1 = int(input("Digite a nota da primeira prova: "))
nota2 = int(input("Digite a nota da segunda prova: "))
media = (nota1 + nota2)/2
print(media)
if media >= 7:
  print("Parabéns! Você foi aprovado")
else:
  print("Você está reprovado")
```

```python
#3 – Um sistema meteorológico quer classificar a sensação térmica do dia. A partir de uma temperatura informada, determine se está fria, agradável, quente ou muito quente usando if…elif…else.
clima = int(input("Digite a sensação térmica do dia: "))
if clima < 20:
  print("O clima está frio")
elif clima >= 20 and clima < 30:
  print("O clima está agradavel")
elif clima >= 30 and clima < 35:
  print("O clima esta quente")
elif clima >= 35:
  print("O clima está muito quente")
```

```python
#5 – Uma empresa deseja reajustar salários com base em faixas de rendimento. Receba o salário atual e aplique um aumento conforme as faixas (10%, 7% ou 5%), usando condicionais encadeadas. Exiba o novo salário.
salario = float(input("Digite o seu salário: "))
if salario < 2000:
  aumento = 0.10
elif salario < 5000:
  aumento = 0.07
else:
  aumento = 0.05

novo_salario = salario + (salario * aumento)
print("Seu novo salário é :", novo_salario)
```

```python
#11 – Uma calculadora avançada deve permitir que o usuário some quantos números quiser. Peça números repetidamente até que o valor 0 seja digitado. Ao final, mostre a soma total.
soma = 0
num = float(input("Digite um número(Se quiser parar digite 0): "))
while num != 0:
  soma += num
  num = float(input("Digite um número(Se quiser parar digite 0): "))

print("A soma dos números é:",soma)
```

```python
#14 – Uma pesquisa está registrando idades. Continue lendo idades até que o usuário digite um valor negativo. Ao final, calcule e exiba a média das idades digitadas.
num = int(input("Digite a Idade: "))
soma = 0
quantidade = 0
while num > 0:
  soma += num
  quantidade += 1
  num = int(input("Digite a Idade: "))

if num == 0:
  print("Idade invalida")
else:
  print("A média das idades é:",soma/quantidade)
```

```python
#15 – Um cronômetro digital deve realizar uma contagem regressiva personalizada. Receba um número inicial e exiba a contagem até 0 usando um while.
n = int(input("Digite um número para realizar a contagem regressiva: "))
while n > 0:
  print(n)
  n -= 1
  if n == 0:
    print(n,"Acabou!")
```

```python
#21 – Um app de produtividade quer identificar qual de quatro tarefas possui o nome mais longo. Leia quatro nomes de tarefas e exiba aquela com mais caracteres.
app = []
for a in range(4):
  n = input("Digite uma palavra: ")
  app.append(n)

maior = app

for n in app:
  if len(n) > len(maior):
    maior = n

print(f"A palavra coma mais caracteres é {maior}")
```

```python
#23 – Um módulo de análise precisa classificar uma lista de valores. Leia 10 números e exiba quantos são positivos, negativos e quantos são iguais a zero.
positivos = 0
negativos = 0
nulo = 0

for n in range(10):
  num = int(input("Digite um número: "))
  if num > 0:
    positivos += 1
  if num < 0:
    negativos += 1
  if num == 0:
    nulo += 1

print(f"A quantidade de números positivos é {positivos}")
print(f"A quantidade de números negativos é {negativos}")
print(f"A quantidade de números igual a 0 é {nulo}")
```

```python

#24 – Uma escola quer filtrar nomes que comecem com determinadas letras. Leia 6 nomes e mostre apenas aqueles que começam com “A” ou “C”.
nomes = []

for i in range(6):
    nomes.append(input("Digite um nome: "))

for nome in nomes:
    if nome[0] in "a" or nome[0] in "A" or nome[0] in "c" or nome[0] in "C":
        print("Nome filtrado:", nome)
```

```python
#25 – Uma caixa registradora digital permite que o usuário digite valores até inserir um código especial. Continue lendo números até que o valor 999 seja digitado. Use break para encerrar.
especial = 999
valor = int(input("Digite um código: "))
while valor != especial:
  valor = int(input("Digite um código: "))
  if valor == especial:
    print("Digitou o Código especial")
    break
```

```python
#27 – Crie um sistema de login com limite de tentativas. O usuário terá no máximo 3 tentativas para digitar a senha correta. Caso acerte antes, use break para encerrar o laço.
senha = "1234"
tentativas = 0

while tentativas < 3:
    num = input("Digite a senha: ")
    tentativas += 1
    if num == senha:
      print("Senha correta")
      break

if tentativas == 3 and num != senha:
    print("Limite de tentativas atingido")
```

```python
#28 – Desenvolva um programa que leia 10 números e some apenas os valores positivos. Caso um número negativo seja digitado, utilize continue para ignorá-lo.
quantidade = 0

for a in range(10):
    num = int(input("Digite um número: "))
    if num < 0:
        continue
    quantidade += num

print(f"A soma dos números positivos é {quantidade}")
```

```python

#29 – Simule um acumulador que percorre números até que a soma ultrapasse 200. Quando isso ocorrer, encerre o laço usando break.
soma = 0
n = 1
while soma < 200:
    soma += n
    n += 1
    print(soma)
    if soma > 200:
      break
```

```python
#30 – Um programa de edição quer contar quantas palavras o usuário digitou. Continue lendo palavras até que o usuário digite “fim”. Depois, exiba quantas palavras válidas foram informadas.
palavras = 0
fim = "fim"
pav = input("Digite uma senha: ")
while pav != fim:
  palavras += 1
  pav = input("Digite uma senha: ")
  if pav == fim:
    break

print(f"A quantidade de palavras que foi digitado foi {palavras}")
```
